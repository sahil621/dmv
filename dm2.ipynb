```python
# Set the location for which you want to retrieve weather data
lat = 18.184135
lon = 74.610764
```

```python
#https://openweathermap.org/api/
#how to use api call
#Construct the APL URL
api_url = f"https://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}"
```

```python
# Send a GET request to the API
res = requests.get(api_url)
weather_data = res.json()
weather_data
```

```python
# Getting the data from dictionary and taking into onevariable
# Extract relevant weather attributes using list
weather_data['list'][0]['weather'][0]['description']
```

```python
temperatures = [item['main']['temp'] for item in weather_data['list']]
temperatures
```

```python
# It will extract all values (40) and putting into one variable
timestamps = [pd.to_datetime(item['dt'], unit='s') for item in weather_data['list']]
timestamps
```

```python
# Create a pandas DataFrame with the extracted weather data
humidity = [item['main']['humidity'] for item in weather_data['list']]
humidity
```

```python
# Set the Timestamp column as the DataFrame's index
weather_description = [item['weather'][0]['description'] for item in weather_data['list']]
weather_description
```

```python
wind_speed = [item['wind']['speed'] for item in weather_data['list']]
wind_speed
```

```python
 # Create a pandas DataFrame with the extracted weather data
df = pd.DataFrame({
    'Timestamp' : timestamps,
    'Temperature' : temperatures,
    'Humidity' : humidity,
    'Wind Speed' : wind_speed,
    'Weather Description' : weather_description
})

df
```

```python
# Set the Timestamp column as the DataFrame's index
df.set_index('Timestamp', inplace=True)
```

```python
df
```

```python
df['Temperature'].dtype
```

```python
# it was in kelvin (temperature)
df['Temperature'] = df['Temperature'].apply(lambda x: x - 273.15)
```

```python
df
```

```python
import matplotlib.pyplot as plt

daily_mean_temp = df['Temperature'].resample('D').mean()
daily_mean_humidity = df['Humidity'].resample('D').mean()
daily_mean_wind_speed = df['Wind Speed'].resample('D').mean()

daily_mean_temp
```

```python
# Plot the mean daily temperature over time(line plot)
plt.figure(figsize=(10, 5))
daily_mean_temp.plot(color="red", linestyle='-')
plt.title("Daily Mean Temperature")
plt.xlabel("Date")
plt.ylabel("Temperature (Celsius)")
plt.grid(True)
plt.show()
```

```python
# Plot the mean daily humidity over time (bar plot)
plt.figure(figsize=(10, 5))
daily_mean_humidity.plot(kind="bar", color="blue")
plt.title("Daily Mean Humidity")
plt.xlabel("Date")
plt.ylabel("Humidity (%)")
plt.grid(True)
plt.show()
```

```python
# Plot the relationship between temperature and wind speed(scatter plot)
plt.figure(figsize=(10, 5))
plt.scatter(df['Temperature'], df['Humidity'], color='green')
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature (Celsius)")
plt.ylabel("Humidity (%)")
plt.grid(True)
plt.show()
```
